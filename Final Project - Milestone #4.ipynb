```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Adoptable_Pets_Cleaned.csv")
```

# Question 1: What types of animals are most common in the shelter?

### Count occurrences of each animal type and display:

```python
animal_counts = df["type"].value_counts()
animal_counts
```

#### Code for Plot:

```python
plt.figure(figsize=(8,5))
sns.barplot(x=animal_counts.index, y=animal_counts.values)
plt.title("Number of Animals by Type")
plt.xlabel("Animal Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Conclusions: 
Cats make up the largest share of animals in the shelter, with nearly 40 cats recorded. Dogs are the second-largest group at around 30. After that, there is a big drop in numbers: only a small handful of “Other” animals appear in the dataset, and birds are extremely rare, with only one entry.

This shows that the shelter primarily works with cats and dogs, which matches what you would normally expect at most animal shelters. Since almost all animals fall into these two categories, most adoption-related trends in the dataset will likely come from cats and dogs rather than the smaller groups.

# Question 2: Does pet size vary by animal type?

### Create a frequency table of animal type vs. size

```python
size_by_type = df.groupby(["type", "size"]).size().unstack(fill_value=0)
size_by_type
```

#### Code for Plot:

```python
plt.figure(figsize=(12,6))
size_by_type.plot(kind="bar", width=0.8, figsize=(12,6))

plt.title("Pet Size Distribution by Animal Type")
plt.xlabel("Animal Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Size")
plt.tight_layout()
plt.show()
```

#### Conclusions: 

The pet size chart shows clear differences between animal types. Cats appear most often in the “Medium” size category, with a noticeable number of “Small” cats as well. Dogs show more variation, with many “Large” and “Medium” dogs, and a few extra-large and small dogs. Birds and “Other” animals are almost always in the “Small” or “Medium” categories.

Overall, this suggests that dogs have the widest range of sizes in the shelter, while cats stay mostly within the medium-to-small range. This is helpful for planning kennel space and resources, since larger dogs require more room and different supplies compared to cats and smaller animals.
